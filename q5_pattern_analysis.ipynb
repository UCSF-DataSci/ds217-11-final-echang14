````python
# part 1: correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
key_variables = ['Air Temperature', 'Humidity', 'Wind Speed', 'Rain Intensity', 'Barometric Pressure', 'Solar Radiation',"Wet Bulb Temperature", "Battery Life", "Total Rain", "Wind Direction"]
numeric_df = numeric_df[key_variables]
corr_matrix = numeric_df.corr()
corr_matrix.to_csv('output/q5_correlations.csv')
print("Saved correlation matrix to output/q5_correlations.csv")

````

````python
# part 2: patterns
fig, axs = plt.subplots(1, 3, figsize=(15, 10))

# 1. Temporal trend (Air Temperature over time)
monthly_avg = df['Air Temperature'].resample('M').mean().dropna()
axs[0].plot(monthly_avg.index, monthly_avg.values, color="teal")
axs[0].set_title("Monthly Average Air Temperature Over Time")
axs[0].set_xlabel("Time")
axs[0].set_ylabel("Air Temperature (C)")

# 2. Seasonal pattern (average air temperature by month)
df['Month'] = df.index.month
monthly_avg = df.groupby('Month')['Air Temperature'].mean().dropna()
axs[1].bar(monthly_avg.index, monthly_avg.values, color="salmon")
axs[1].set_title("Average Air Temperature by Month")
axs[1].set_xlabel("Month")
axs[1].set_ylabel("Air Temperature (C)")

# 3. Correlation heatmap
sns.heatmap(corr_matrix, ax=axs[2], cmap="coolwarm", center=0)
axs[2].set_title("Feature Correlation Heatmap")



plt.suptitle("Key Patterns in Environmental Data", fontsize=16)
plt.tight_layout()
plt.savefig("output/q5_patterns.png", dpi=150)
````
